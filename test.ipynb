```python
# save metadata as an excel file
df2007 = pd.DataFrame(list(meta2007.column_names_to_labels.items()), columns=['Variable', 'Label'])
df2009 = pd.DataFrame(list(meta2009.column_names_to_labels.items()), columns=['Variable', 'Label'])
df2011 = pd.DataFrame(list(meta2011.column_names_to_labels.items()), columns=['Variable', 'Label'])
df2013 = pd.DataFrame(list(meta2013.column_names_to_labels.items()), columns=['Variable', 'Label'])
df2015 = pd.DataFrame(list(meta2015.column_names_to_labels.items()), columns=['Variable', 'Label'])

met_dict_2007 = pd.DataFrame.from_dict(meta2007.column_names_to_labels, orient='index', columns=['Label'])
met_dict_2009 = pd.DataFrame.from_dict(meta2009.column_names_to_labels, orient='index', columns=['Label'])
met_dict_2011 = pd.DataFrame.from_dict(meta2011.column_names_to_labels, orient='index', columns=['Label'])
met_dict_2013 = pd.DataFrame.from_dict(meta2013.column_names_to_labels, orient='index', columns=['Label'])
met_dict_2015 = pd.DataFrame.from_dict(meta2015.column_names_to_labels, orient='index', columns=['Label'])

# Replacing the column names with the labels
data_2007.columns = [met_dict_2007.loc[x][0] for x in data_2007.columns]
data_2009.columns = [met_dict_2009.loc[x][0] for x in data_2009.columns]
data_2011.columns = [met_dict_2011.loc[x][0] for x in data_2011.columns]
data_2013.columns = [met_dict_2013.loc[x][0] for x in data_2013.columns]
data_2015.columns = [met_dict_2015.loc[x][0] for x in data_2015.columns]

# save the data as an excel file
data_2007.to_excel('./xlsx with col replaced/data_2007_col_replaced.xlsx', index=False)
data_2009.to_excel('./xlsx with col replaced/data_2009_col_replaced.xlsx', index=False)
data_2011.to_excel('./xlsx with col replaced/data_2011_col_replaced.xlsx', index=False)
data_2013.to_excel('./xlsx with col replaced/data_2013_col_replaced.xlsx', index=False)
data_2015.to_excel('./xlsx with col replaced/data_2015_col_replaced.xlsx', index=False)
```
